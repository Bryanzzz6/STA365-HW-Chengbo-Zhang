```python
#Q2-1

import pymc as pm
import numpy as np
import arviz as az
import matplotlib.pyplot as plt

# Generate sample data
n = 30
x = np.random.normal(loc=0, scale=1, size=n)

# Bayesian inference with Normal prior for theta and Gamma prior for tau
with pm.Model() as model1:
    theta = pm.Normal("theta", mu=0, sigma=1)  # Normal prior for theta
    tau = pm.Gamma("tau", alpha=2, beta=1)  # Gamma prior for precision (1/sigma^2)

    # Convert precision to standard deviation
    sigma = pm.Deterministic("sigma", 1 / pm.math.sqrt(tau))

    # Likelihood function
    x_obs = pm.Normal("x_obs", mu=theta, sigma=sigma, observed=x)

    # MCMC sampling
    idata1 = pm.sample(1000, chains=2, target_accept=0.9, return_inferencedata=True)

# Diagnostic plots for Model 1
az.plot_trace(idata1)
plt.show()

# Print summary statistics (including R-hat for convergence)
print(az.summary(idata1))

print(az.summary(idata1, hdi_prob=0.94))
az.plot_posterior(idata1, hdi_prob=0.94)
plt.show()
```

### 1. How the Markov Algorithm of Slice Sampling Works

Slice sampling is a Markov Chain Monte Carlo (MCMC) method used to sample from complex distributions when the normalization constant is unknown. The core idea is to introduce an **auxiliary variable** and sample uniformly from the region under the probability density curve, known as the "slice."

#### Steps in the Algorithm

1. Define the Target Function $( f(x))$  
This is the probability density function (PDF) we want to sample from (known up to a normalizing constant).


2. Initialization:  
Start with an initial value $( x_0)$.


3. Vertical Sampling:  
Draw a uniform variable $ y \sim U(0, f(x_0)) $ to define the horizontal slice beneath the curve.


4. Horizontal Sampling (Slice Sampling):  
Identify an interval $( (L, R) )$ where $( f(x) > y)$, initially centered around $( x_0)$.
Expand the interval outward if needed to ensure it fully contains the slice.
Draw a new candidate $( x^* \sim U(L, R))$.


5. Acceptance Step:
If $( f(x^*) > y )$, accept $( x^*)$.
Otherwise, shrink the interval and re-sample until acceptance.

Why Slice Sampling is More Efficient than Metropolis-Hastings:
Automatically adapts to the shape of the target density → avoids rejection tuning.
Works well when the target distribution is only known up to a normalizing constant.
No need for a proposal distribution → avoids inefficient tuning of step sizes.


### 2. Replacing Metropolis-Hastings in Gibbs Sampling

When using a Metropolis within Gibbs sampler, if the full conditional distributions are known only up to a normalizing constant, slice sampling can be a robust alternative.

#### Key Differences

| Method | How it Works |
|--------|-------------|
| Metropolis-Hastings | Proposes a new value and accepts/rejects based on acceptance probability. |
| Slice Sampling | Finds a slice under the full conditional and samples directly from it. |

#### Steps for Gibbs Sampling with Slice Sampling

1. Full Conditional Sampling:  
Instead of proposing a new state and accepting/rejecting, slice sampling directly samples from the slice defined by the current state.

2. Defining the Slice:  
The "curve" we sample beneath is the unnormalized full conditional density.

3. Initial Value: 
The current state of the Markov chain serves as the starting point.

4. Steps to Create a Draw for Gibbs Sampling:

Step 1: Sample $( y \sim U(0, f(x_{\text{current}})) )$.
   
Step 2: Find an interval $( (L, R) )$ such that $( f(x) > y )$.
   
Step 3: Sample $( x^* \sim U(L, R) )$.
   
Step 4: Accept $( x^* )$ if $( f(x^*) > y)$, otherwise shrink the interval and repeat.
   
Step 5: Repeat for each variable in the Gibbs cycle.

#### Potential Issue
 If the full conditional is multi-modal, slice sampling may struggle to move between modes.



### 3. Explanation of Key Concepts

#### What is the Curve We're Sampling Beneath?
The target function $( f(x) )$ is the unnormalized full conditional density.
The slice at height $( y )$ defines a horizontal region from which we sample $( x )$.

#### What is the Initial Value?
The current sample $( x_i )$ from the previous Gibbs iteration.

#### What are the Steps to Create a Draw for Gibbs Sampling?
Fix other parameters.
Define the full conditional for $( x_i )$.
Use Slice Sampling draw $( y )$, expand interval, sample $( x )$.
Update $( x_i )$ and proceed to the next variable.
